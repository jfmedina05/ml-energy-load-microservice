```python
# Import necessary Library
import pandas as pd

# Makes the base, non optimized data the base data tested
base_data = pd.read_csv('ENB2012_data.csv')
```

### Split Data into train/test set

```python
# Rename columns to better understand the data
base_data.rename(columns={
    "X1": "Relative Compactness",
    "X2": "Surface Area",
    "X3": "Wall Area",
    "X4": "Roof Area",
    "X5": "Overall Height",
    "X6": "Orientation",
    "X7": "Glazing Area",
    "X8": "Glazing Area Distribution",
    "Y1": "Heating Load",
    "Y2": "Cooling Load"
}, inplace=True)

# Display the head of the data to see if renaming worked
base_data.head()
```

```python
# Import neccesary Library
from sklearn.model_selection import train_test_split

# Prepare features and target
X = base_data.drop(columns=['Heating Load', 'Cooling Load'])
Y = base_data[['Heating Load', 'Cooling Load']]

# Split into train and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

### Load data from optimized set

```python
# Import the necessary Library
from joblib import dump, load
op_mdl = load('rfr.pkl')
```

### Predictions for test data

```python
# Predict test data
y_pred = op_mdl.predict(X_test)
print(y_pred)
```

### Performance Metrics

```python
# Import necessary library
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Calculate performance metrics
mae = mean_absolute_error(Y_test, y_pred)
mse = mean_squared_error(Y_test, y_pred)
r2 = r2_score(Y_test, y_pred)

# Print results
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"R² Score: {r2:.4f}")
```

### Save the test data

```python
# Save test dataset
test_data = X_test.copy()
test_data[['Heating Load', 'Cooling Load']] = Y_test
test_data.to_csv("test_data.csv", index=False)

print("Test data saved as test_data.csv")
```
